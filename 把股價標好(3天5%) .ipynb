```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('大立光2023-2025股價表 - 工作表1.csv')
df.head(5)
```

```python
df_date_roi = df[['年月日' ,'三日報酬率']]
df_date_roi
```

```python
df_date_roi['三日報酬率'] = df_date_roi['三日報酬率'].str.replace('%', '').astype(float)
df_date_roi['年月日'] = pd.to_datetime(df_date_roi['年月日'])
df_date_roi.info()
```

```python
#迭代
df_date_roi.shape[0]
df_date_roi['當天文章類別'] = ''
for i in range(df_date_roi.shape[0] - 3):
    roi = df_date_roi['三日報酬率'][i]
    if roi >= 5:
        df_date_roi.at[i+3, '當天文章類別'] = '漲'
    elif roi <= -5:
        df_date_roi.at[i+3, '當天文章類別'] = '跌'
    else:
        df_date_roi.at[i+3, '當天文章類別'] = '平'
df_date_roi = df_date_roi.iloc[3:514] #因為前面三個沒有後面的日子幫忙判斷
df_date_roi
        
```

```python
df_date_roi['當天文章類別'].value_counts()
```

把兩個資料集合併

```python
df_news = pd.read_csv('df_largon_with_date_news.csv')
df_news.head(5)
```

```python
df_news.info()
```

```python
df_news['post_time'] = pd.to_datetime(df_news['post_time']).dt.date
df_news.head(5)
```

```python
df_ppt = pd.read_csv('df_largon_with_date_ppt.csv')
df_ppt.head(5)
```

```python
df_ppt.info()
```

```python
df_ppt['title'] = df_ppt['title'].str.replace('[新聞]', '', regex=False)
df_ppt['post_time'] = pd.to_datetime(df_ppt['post_time']).dt.date
df_ppt.head(5)
```

```python
combined_df = pd.concat([df_ppt, df_news], ignore_index=True)
combined_df.shape
```

```python
#開始對照
# 建立對照表：用年月日對應類別
class_mapping = df_date_roi.set_index('年月日')['當天文章類別']

# 使用 .map() 對齊日期，對照文章類別
combined_df['class'] = combined_df['post_time'].map(class_mapping)

combined_df.head(5)
```

```python
combined_df['class'].value_counts()
```

文章分類(漲)

```python
#先分漲
df_3d5_up = combined_df[combined_df['class']=='漲']
df_3d5_up.shape
```

```python
#漲的只留title跟content 
df_3d5_up = df_3d5_up[['title','content']]
df_3d5_up
```

```python
#儲存成csv檔
df_3d5_up.to_csv('3天5%的資料集/df_3d5_up.csv', index=False, encoding='utf-8-sig')
```

文章分類(跌)

```python
#分跌
df_3d5_down = combined_df[combined_df['class']=='跌']
df_3d5_down.shape
```

```python
#只留title跟content 
df_3d5_down = df_3d5_down[['title','content']]
df_3d5_down
```

```python
#儲存成csv檔
df_3d5_down.to_csv('3天5%的資料集/df_3d5_down.csv', index=False, encoding='utf-8-sig')
```
